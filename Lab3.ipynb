```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv("WorldEnergy.csv")

countries = ['China', 'United States', 'India', 'Germany', 'Brazil']
df = df[df['country'].isin(countries)]

df = df[['country', 'year', 'energy_per_capita']].dropna()

print("Sample size by country:")
print(df['country'].value_counts())
print()
```

```python
plt.figure(figsize=(10, 6))
data_to_plot = [df[df['country'] == c]['energy_per_capita'] for c in countries]
plt.boxplot(data_to_plot, labels=countries)
plt.title('Distribution of Energy Consumption per Capita by Country')
plt.xlabel('Country')
plt.ylabel('Energy per Capita')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
mean_values = df.groupby('country')['energy_per_capita'].mean().reindex(countries)

plt.figure(figsize=(10, 6))
plt.bar(mean_values.index, mean_values.values)
plt.title('Mean Energy Consumption per Capita by Country')
plt.xlabel('Country')
plt.ylabel('Mean Energy per Capita')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for c in countries:
    subset = df[df['country'] == c]
    yearly_mean = subset.groupby('year')['energy_per_capita'].mean()
    plt.plot(yearly_mean.index, yearly_mean.values, label=c)

plt.title('Energy Consumption per Capita Over Time')
plt.xlabel('Year')
plt.ylabel('Energy per Capita')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# H0: Mean energy_per_capita is equal across countries
# H1: At least one country has a different mean
# -----------------------------
groups = [df[df['country'] == c]['energy_per_capita'] for c in countries]

f_stat, p_value = stats.f_oneway(*groups)

print("ANOVA Result")
print(f"F-statistic = {f_stat:.4f}")
print(f"p-value = {p_value:.6f}")
print()

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H0")
    print("There is a statistically significant difference in mean energy consumption per capita among the selected countries.")
else:
    print("Conclusion: Fail to reject H0")
    print("There is no statistically significant difference in mean energy consumption per capita among the selected countries.")
```
